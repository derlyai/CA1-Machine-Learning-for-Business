Se podria optar por rellenar los valores faltantes con la media o mediana, pero el enfoque kNN tiende a dar mejores resultados, especialmente cuando las características tienen relaciones no triviales. Por esta razon se decide utilizer un metodo no supervisado.

```python
from sklearn.impute import KNNImputer
import pandas as pd
```

```python
impute = KNNImputer(n_neighbors=5)
```

```python
data_numer = data.drop(columns=['ocean_proximity'])
```

```python
df_imputed = impute.fit_transform(data_numer)
```

```python
df_imputed = pd.DataFrame(df_imputed, columns=data_numer.columns)
```

```python
df_imputed['ocean_proximity'] = data['ocean_proximity']
```
